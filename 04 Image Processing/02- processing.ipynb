```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

---

### `1. Low Pass Filter (Smoothing and Noise Reduction)`
* `Low frequencies` in an image are characterized by slow or gradual changes in pixel intensity, which represent smooth areas of the image.
* `High frequencies` signify rapid changes in intensity, such as edges or noise.

> * `Low-pass filters smooth an image by averaging the pixels with their neighbors.` <br />
> * `They are often used for noise reduction and image blurring.` <br/>
> * `The sum of all the values in a low pass filter is typically equal to 1.` <br />
> * `Common examples include the averaging filter and Gaussian filter.`

* `Using Average Filter`

```python
# Load an image
image = cv2.imread(filename='./basha.jpg')

# Define the averaging filter
kernel_size = 5  # Size of the kernel
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)

# Apply the averaging filter
smoothed_image = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)  ## ddepth (destination depth) -> A value of -1 means that the depth will be the same as the source.

# Display the original and smoothed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Smoothed Image')
plt.imshow(cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

* `Using Gaussian Filter`

```python
# Load an image
image = cv2.imread(filename='./basha.jpg')

# Apply Gaussian blur
gaussian_blur = cv2.GaussianBlur(src=image, ksize=(5, 5), sigmaX=0)  # sigmaX - > std of Gaussian Kernel

# Display the original and Gaussian blurred images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(src=image, code=cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gaussian Blurred Image')
plt.imshow(cv2.cvtColor(src=gaussian_blur, code=cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

### `2. High Pass Filter (Edge Detection and Sharpening)`
> * `High pass filters are used to enhance the edges and details in an image by removing the low-frequency components.`
> * `The sum of all values in a high pass filter is typically equal to 0.`
> * `Common examples include the Sobel filter and Laplacian filter.`

* `Using Sobel Filter`

```python
# Load an image
image = cv2.imread(filename='./basha.jpg', flags=cv2.IMREAD_GRAYSCALE)

# Apply Sobel filter
sobel_x = cv2.Sobel(src=image, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)  # Edge detection in x direction
sobel_y = cv2.Sobel(src=image, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)  # Edge detection in y direction
# ddepth=cv2.CV_64F --> indicates the depth of the output image. 
# CV_64F --> is used here to handle negative values that can arise during the gradient calculation.

# Combine the two Sobel images
sobel_combined = cv2.magnitude(x=sobel_x, y=sobel_y)

# Display the Sobel filtered images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel X Image')
plt.imshow(sobel_x, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sobel Combined Image')
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')

plt.show()
```

* `Using Laplacian Filter`

```python
# Load an image
image = cv2.imread(filename='./basha.jpg', flags=cv2.IMREAD_GRAYSCALE)

# Apply Laplacian filter
laplacian = cv2.Laplacian(src=image, ddepth=cv2.CV_64F, ksize=7)

# Display the original and Laplacian filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Laplacian Image')
plt.imshow(laplacian, cmap='gray')
plt.axis('off')

plt.show()
```
